```python
def calculate_expected_profit(multiplier, hunters, total_players, total_expeditions, expedition_cost):
    base_treasure = 7500
    total_hunters = hunters + 1  # Adding 1 for myself
    percentage = total_hunters / total_players
    total_hunters_and_percentage = total_hunters + percentage
    treasure = base_treasure * multiplier
    profit = treasure / total_hunters_and_percentage - expedition_cost
    return profit

data = {
    'Tile': ['G26', 'G27', 'G28', 'G29', 'G30', 'H26', 'H27', 'H28', 'H29', 'H30', 'I26', 'I27', 'I28', 'I29', 'I30', 'J26', 'J27', 'J28', 'J29', 'J30', 'K26', 'K27', 'K28', 'K29', 'K30'],
    'Multiplier': [24, 70, 41, 21, 60, 47, 82, 87, 80, 35, 73, 89, 100, 90, 17, 77, 83, 85, 79, 55, 12, 27, 52, 15, 30],
    'Hunters': [2, 4, 3, 2, 4, 3, 5, 5, 5, 3, 4, 5, 8, 7, 2, 5, 5, 5, 5, 4, 2, 3, 4, 2, 3]
}

total_players = 5000 # Number of teams in estimate
total_expeditions = 1  # First expedition is free
expedition_costs = [0, 25000, 75000]  # Costs for second and third expeditions

# Calculate expected profit for each destination for each expedition
for i, cost in enumerate(expedition_costs):
    print(f"Expected profit for Expedition {i+1} (Cost: {cost} SeaShells):")
    
    # Calculate expected profit for each destination
    profits = []
    for tile, multiplier, hunters in zip(data['Tile'], data['Multiplier'], data['Hunters']):
        profit = calculate_expected_profit(multiplier, hunters, total_players, total_expeditions, cost)
        profits.append((tile, profit))
    
    # Sort destinations based on expected profit in descending order
    sorted_profits = sorted(profits, key=lambda x: x[1], reverse=True)

    for tile, profit in sorted_profits:
        print(f"Destination: {tile}, Expected Profit: {profit:.2f} SeaShells")
    
    print()
```
